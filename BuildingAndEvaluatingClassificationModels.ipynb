```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine_data = pd.read_csv('datasets/winequality-red.csv')

wine_data.head()
```

```python
wine_data.shape
```

```python
sorted(wine_data['quality'].unique())
```

```python
bins = (2, 5.5, 8)
labels = ['mediocre', 'good']

wine_data['categories'] = pd.cut(x = wine_data['quality'], bins = bins, labels=labels )
```

```python
wine_data.head(10)
```

```python
wine_data['categories'].unique()
```

```python
wine_data['categories'].value_counts()
```

```python
wine_data['categories']= np.where(wine_data['categories'] == 'mediocre',0,1)
wine_data.head()
```

```python
wine_data.drop('quality', inplace=True, axis=1)
```

```python
wine_data.head()
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(wine_data['alcohol'], wine_data['density'])

plt.title('scatter plot for alcohol vs Density', fontsize=20)
plt.xlabel('alcohol', fontsize=12)
plt.ylabel('density', fontsize=12)
plt.show()
```

```python
wine_data.boxplot(by='categories',
                 column = 'alcohol',
                 figsize=(12,8), fontsize=15, grid= False)

plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.violinplot(wine_data['alcohol'])

plt.xlabel('Alcohol')
plt.show()
```

```python
wine_data.to_csv('datasets/winequality-red-processed.csv', index=False)
```
